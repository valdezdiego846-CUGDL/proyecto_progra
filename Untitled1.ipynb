```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Iniciando análisis de Creatividad Digital...\n")

# 1. Cargar la base de datos (Directo desde Colab)
try:
    df = pd.read_csv('campanas_creatividad_digital.csv')
    print("✅ Datos cargados correctamente. Total de campañas:", len(df))
except FileNotFoundError:
    print("❌ ERROR: No se encontró el archivo 'campanas_creatividad_digital.csv'. Asegúrate de haberlo subido a Colab.")

# 2. Creación de Variables Derivadas
df['roi'] = (df['ingresos_usd'] - df['costo_usd']) / df['costo_usd']
df['ctr'] = (df['clicks'] / df['impresiones']) * 100
df['cvr'] = (df['conversiones'] / df['clicks']) * 100
df['cpc'] = df['costo_usd'] / df['clicks']

# 3. Hallazgos Estadísticos
mejor_plataforma = df.groupby('plataforma')['roi'].mean().idxmax()
mejor_formato = df.groupby('formato')['ctr'].mean().idxmax()

top_5_roi = df.nlargest(5, 'roi')[['id_campana', 'plataforma', 'formato', 'roi']]
bottom_5_roi = df.nsmallest(5, 'roi')[['id_campana', 'plataforma', 'formato', 'roi']]

# 4. Detección de Outliers (IQR en CTR)
Q1 = df['ctr'].quantile(0.25)
Q3 = df['ctr'].quantile(0.75)
IQR = Q3 - Q1
limite_superior = Q3 + 1.5 * IQR
limite_inferior = Q1 - 1.5 * IQR

outliers_ctr = df[(df['ctr'] > limite_superior) | (df['ctr'] < limite_inferior)]

# --- IMPRESIÓN DE RESULTADOS ---
print("-" * 50)
print(f"🏆 Mejor plataforma por ROI: {mejor_plataforma}")
print(f"🎯 Mejor formato por CTR: {mejor_formato}")
print(f"⚠️ Número de outliers detectados en CTR: {len(outliers_ctr)}")
print("-" * 50)
print("\n🟢 TOP 5 Campañas por ROI:")
print(top_5_roi.to_string(index=False))
print("\n🔴 BOTTOM 5 Campañas por ROI:")
print(bottom_5_roi.to_string(index=False))
print("-" * 50)

# 5. Generación de Gráficas
sns.set_theme(style="whitegrid")

# Gráfica 1: Boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(x='plataforma', y='roi', data=df, palette='Set2', hue='plataforma', legend=False)
plt.title('Distribución de ROI por Plataforma', fontsize=14)
plt.xlabel('Plataforma')
plt.ylabel('ROI')
plt.tight_layout()
plt.show()

# Gráfica 2: Bar chart
plt.figure(figsize=(10, 5))
ctr_promedio = df.groupby('formato')['ctr'].mean().sort_values(ascending=False).reset_index()
sns.barplot(x='formato', y='ctr', data=ctr_promedio, palette='viridis', hue='formato', legend=False)
plt.title('CTR Promedio por Formato de Anuncio', fontsize=14)
plt.xlabel('Formato')
plt.ylabel('CTR Promedio (%)')
plt.tight_layout()
plt.show()

# Gráfica 3: Dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='costo_usd', y='ingresos_usd', hue='plataforma', style='formato', data=df, s=100, palette='deep')
plt.title('Costo vs Ingresos (Agrupado por Plataforma y Formato)', fontsize=14)
plt.xlabel('Costo de Campaña (USD)')
plt.ylabel('Ingresos Generados (USD)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("\n✅ Análisis completado con éxito.")
```
